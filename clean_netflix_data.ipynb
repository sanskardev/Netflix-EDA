```python
import pandas as pd
```

```python
df = pd.read_csv('/content/netflix_titles.csv')
df
```

remove \n from title

```python
print(df[df['title'].str.contains('\n')]['title'])
```

```python
df['title'] = df['title'].str.replace('\n', ' ')
```

```python
print(df[df['title'].str.contains('Memphis')]['title'])
```

remove \n from description

```python
print(df[df['description'].str.contains('\n')]['description'])
```

```python
df['description'] = df['description'].str.replace('\n', ' ')
```

```python
df.to_csv('netflix_title_clean.csv', index=False)
```
